# HTML
**Hyper Text Markup Language**
- HTML이란?
  - 웹 페이지를 작성하기 위한 역할 표시 언어 (마크업 언어)
    - HTML 태그를 통해 정보를 구조화함으로써 웹 페이지의 내용과 구조를 담당한다.
    - 역할을 나타낸다 == 구조를 잘 잡는다
  - Hyper의 의미
    - 선형적으로 텍스트를 찾던 기존의 방식에서 나아가, 다양한 갈래로 텍스트를 찾을 수 있게 되면서 시간을 대폭 단축시킬 수 있게 되었다.
- HTTP란?
  - Hyper Text Transfer Protocol
  - Hyper Text를 주고받는 규칙, 즉 전송하는 통신 규약
- 마크업이란?
  - 역할 표시를 수행한다.
  - 마크업의 간략한 기능을 수행하는 것이 바로 마크다운이다.

## 태그
- 파이썬과 다르게 반드시 쌍따옴표를 구분해야한다.
- 마크다운과 동일하게 엔터 및 띄어쓰기를 여러번 입력하더라도 한번만 적용된다.
- 모든 태그가 클로징이 있는 것은 아니다.
  - `<br>`, `<hr>`, `<input>`, `<img>` 등
- 요소 * number 적용 시 숫자만큼 반복 출력된다.

1. **기초**
    `<!DOCTYPE html>`  html 작성을 시작하겠다는 의미
    
    `<>` 오프닝 태그
    
    `</>` 클로징 태그
    
    `ctrl+enter`: 오프닝, 클로징 사이에 줄바꿈 안되고 바로 다음줄로 내려가는 단축키
    
    `alt+B`  open in default browser (vscode에서 파일 설치해서 가능함)
    
    `ctrl+/`  주석
    
    `emmet`  html을 쉽게 쓰도록 도와주며, `!` 느낌표를 쓰고 `tab`을 누르면 기본 설정값이 나타난다.
    
    `utf-8`  가장 많이 쓰이는, 문자열을 숫자로 치환하는 인코딩 방법
    
    `<head></head>` 사이에는 메타데이터 등이 저장됨
      - `<meta content="width=device-width, initial-scale=1.0">`
        가로길이: 모니터 가로길이, 화면비율: 1배 적용 코드
    

2. **Text**
    `<h1></h1>` 마크다운의 `#`과 동일한 역할
      - <h1> ~ <h6>
    
    `<b></b>` `<strong></strong>`  굵게
    
      - bold는 외관 및 스타일을 나타낸다.
      - strong은 역할 및 목적을 나타낸다.
      - 태그의 기능은 같지만 목적이 다르다.
      - 컴퓨터는 스타일보단 역할과 목적의 중요도를 더 높게 인식한다.
      
    `<i></i>` `<em></em>`  이탤릭
    
      - italic은 외관 및 스타일을 나타낸다.
      - emphasize는 역할 및 목적을 나타낸다.
      - 태그의 기능은 같지만 목적이 다르다.
      - 컴퓨터는 스타일보단 역할과 목적의 중요도를 더 높게 인식한다.
    
        ###### SEO (Search Engine Optimization)
        - 검색 엔진 최적화
          - bold와 strong, italic과 emphasize 구분 등의 같이, 컴퓨터의 입맛에 맞춰 웹 사이트가 검색 결과에 더욱 잘 노출되도록 최적화하는 과정
        - 검색 엔진별로 SEO 가이드가 별도 존재한다.
          - 다만 몇 가지 기본 조건들로는, (1)상기 구조화를 잘 만들어내는 Semantic HTML, (2)이미지와 텍스트 사용 등이 있다.


3. **Content**
    `<br>`  line break 줄바꿈
      - 주소 입력 등의 확실한 구분이 필요할 때 사용하며, 문단 사이의 여백을 두기 위한 용도로 남용되면 안된다.
      - 이미지는 기본적으로 줄글과 동일하게 인지되기 때문에, 글과 나란히 붙어있는 형태로 출력된다. 줄바꿈이 필요하다면 <p>, <br> 태그를 사용해야 한다.
    
    `<p></p>`  단락 구분 (paragraph)
    
    `<div></div>`  블록을 만들어서 영역 구분 (division)
    
    `<pre></pre>`  코드의 원시적인 구조를 그대로 출력


4. **link**
    (attribute) `<a href=" ~ ">Naver</a>`  기존창에서 열기
    
    (attribute) `<a href=" ~ " target="_blank">Naver</a>`  새탭에서 열기
    
    `<a href="#">목적지 미정</a>`  목적지 미정이지만 우선 링크 생성


5. **list**
    `<ol></ol>`  순서 있는 리스트
    
    `<ul></ul>`  순서 없는 리스트
    
    `<li></li>`  <ol> 혹은 <ul> 안에 내용을 리스팅하는 역할


6. **table**
    `<table></table>`  테이블 생성
      - table 태그 안에는 `<thead></thead>` & `<tbody></tbody>`가 있다.
      - 한 줄은 tr(table row) == `<tr></tr>`
      - 한 줄 안에 들어가는 각각의 내용은 td(table data) == `<td></td>`


7. **media**
    `<img src="./html5.png" alt="html5 logo" height="300">`
      - 직접 가지고 있는 로컬 이미지 파일의 위치 작성
      - alt는 대체 텍스트로, 시각 장애인에게 읽어줄 이미지 내용 삽입
    
    `<img src="https://coregenicsoftwares.com/wp-content/uploads/2022/01/clipart2065222-1.png" alt="css3 logo" height="300">`
      - 인터넷에서 가지고 오는 이미지 파일의 주소 작성
      - height, width로 이미지 사이즈를 조정할순 있지만, CSS 사용 권장
    
    <iframe width="560" height="315" src=" ~ " title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture; web-share" allowfullscreen></iframe>
      - video 태그가 따로 존재하나, 거의 iframe을 많이 사용
      - 유튜브 공유 > 퍼가기 > iframe 주소


9. **form**
    `<form></form>`  양식
    
    `<form action=" ~ " method="POST">`
      - action == form 태그 아래 데이터를 보낼 목적지 url

# CSS
**Cascading Style Sheet**
- CSS란?
  - HTML 등의 마크업 언어로 작성된 문서가 실제로 웹사이트에 표현되는 방법을 정해주는 스타일 시트 언어
  - (참고) 기본적으로 HTML은 워드와 같은 형태로 만들어졌다.

## Box Model
`box`  교실
  - box를 이동하기 위해서는 margin 값을 설정해야 한다.
  - 예시: 
    - 우측 하단 이동은 margin-left, margin-top
    - 왼쪽과 위쪽에 여백을 주니 밀려서 우측 하단으로 이동하게 된다.
`content`  책상 위 교과서
  - box와 content의 크기는 기본적으로 동일하다.
  - 여백을 주기 위해서는 padding 값을 설정해야 한다.
`border`  책상의 경계선
  - dotted, dashed, solid, double 등등
  - 상하좌우 모두 값 설정 가능하다 (모서리도 깎을 수 있다)
`padding`  border 안의 교과서 외 여백
  - 별도로 지정하지 않으면 padding은 없는 값이다.
`margin`  다른 책상까지의 거리 여백
  - margin 상하좌우 4개의 값은 시계 방향으로 돌아간다.
  - margin 값을 2개만 줄 때는 (a, b) = 상하, 좌우
  - margin 값을 1개만 줄 때는 상하좌우 모두 같다.

- Inline
